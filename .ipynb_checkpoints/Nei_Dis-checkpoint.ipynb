```python
from utils import load_dataset
from torch_geometric.data import Data
from datasets import Dataset
import pandas as pd
from torch_geometric.datasets import WebKB,WikipediaNetwork,Actor,Planetoid
import torch

from torch_geometric.utils import remove_self_loops, add_self_loops
hetero_graphs = ['minesweeper', 'questions', 'roman_empire', 'tolokers', 'amazon_ratings',
                     'squirrel_filtered_directed', 'texas_4_classes', 'chameleon_filtered_directed']
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'

def get_nei_distri(y, adj):
    nei_dis = torch.matmul(adj, y)
    # print(nei_dis)
    # nei_dis = torch.matmul(adj, nei_dis)
    nei_dis = nei_dis
    nei_dis_sum = nei_dis.sum(dim=1, keepdim=True)
    nei_dis = torch.where(nei_dis_sum != 0, nei_dis / nei_dis_sum, torch.zeros_like(nei_dis))
    # nei_dis = nei_dis + one_hot
    return nei_dis

df = pd.DataFrame(columns=['data_name', 'nd', 'intra', 'inter'])

datasets = ['texas', 'wisconsin', 'cornell', 'squirrel', 'chameleon','actor', 'cora', 'citeseer', 'pubmed' ]
# datasets = ['minesweeper', 'questions', 'roman_empire', 'tolokers', 'amazon_ratings']
result = []
for data_name in datasets:
    dataset, data = load_dataset(data_name, device)
    c = dataset.num_classes
    edge_index = data.edge_index
    # edge_index, _ = remove_self_loops(edge_index)
    # edge_index, _ = add_self_loops(edge_index)
    adj = torch.sparse_coo_tensor(edge_index, torch.ones_like(edge_index[0]), size=[data.num_nodes, data.num_nodes]).float()
    y = torch.nn.functional.one_hot(data.y).float()
    nei_distri = get_nei_distri(y, adj)
    intra_dis = torch.zeros((c, 1))
    inter_dis = torch.zeros((c, c))
    for label in range(c):
        nei_distri_label = nei_distri[data.y == label]
        inter_dis[label] = torch.mean(nei_distri_label, dim=0)
        # intra_nd = torch.mean(torch.sum(nei_distri_label ** 2, dim=1) - 1 / c)
        # intra_dis.append(intra_nd * (c /(c-1)))
        matrix = torch.nn.functional.cosine_similarity(nei_distri_label.unsqueeze(1), nei_distri_label.unsqueeze(0), dim=-1)
        matrix = (matrix+1)/2
        intra_dis[label] = torch.mean(matrix.view(-1)) * (nei_distri_label.size(0) / data.num_nodes)
    
    intra = torch.sum(intra_dis, dim=0)
    # inter_dis_sum = torch.sum(inter_dis, dim=0, keepdim=True)
    # inter_dis = torch.where(inter_dis_sum > 0, inter_dis / inter_dis_sum, torch.zeros_like(inter_dis))
    # # print(inter_dis)
    # inter = torch.mean(torch.sum(inter_dis**2, dim=1)-(1 / c))*(c/(c-1))
    print(inter_dis)
    singular_values = torch.linalg.svdvals(inter_dis)
    normalize_singular = singular_values / torch.sum(singular_values)
    be_sum = normalize_singular * torch.log(normalize_singular)
    to_sum = torch.nan_to_num(be_sum, 0)
    inter =  -torch.sum(to_sum) / torch.log(torch.tensor(c))
    
    nd = (intra + inter)/2
    # nd = intra * inter
    result.append(nd)
    
    new_row = pd.DataFrame({'data_name': [data_name], 'nd': [nd.item()], 'intra': [intra.item()], 'inter': [inter.item()]})
    df =  pd.concat([df, new_row], ignore_index=True)
    
pd.options.display.float_format = '{:.2f}'.format
print(df.to_string(index=False, justify='left', col_space=12))
```

```python
datasets = ['texas', 'wisconsin', 'cornell', 'squirrel', 'chameleon','actor', 'cora', 'citeseer', 'pubmed' ]
from utils import load_dataset
def node_homophily(data):
    homophily = torch.zeros(data.num_nodes)
    for i in range(data.num_nodes):
        neighbors = data.edge_index[0][data.edge_index[1] == i].tolist()
        if len(neighbors) <= 0:
            homophily[i] = 0
            continue
        homo = 0
        for j in range(len(neighbors)):
            if data.y[i] == data.y[neighbors[j]]:
                homo += 1
        homophily[i] = homo / len(neighbors)
    homo_full = torch.mean(homophily)
    
    return homo_full.numpy().tolist()

def edge_homophily(data):
    """计算数据集的节点同质性"""
    edge_index = data.edge_index
    y = data.y
    
    # 初始化计数器
    same_label_count = 0
    total_edges = edge_index.size(1)
    
    for i in range(total_edges):
        src, dst = edge_index[:, i]
        if y[src] == y[dst]:
            same_label_count += 1
    
    # 计算同质性
    homophily = same_label_count / total_edges
    return homophily

def neighbor_homophily(data, c):
    edge_index = data.edge_index
    adj = torch.sparse_coo_tensor(edge_index, torch.ones_like(edge_index[0]), size=[data.num_nodes, data.num_nodes]).float()
    y = torch.nn.functional.one_hot(data.y).float()
    nei_distri = get_nei_distri(y, adj)
    neighbor_homo = 0
    for label in range(c):
        nei_distri_label = nei_distri[data.y == label]
        
        singular_values = torch.linalg.svdvals(nei_distri_label)
        normalize_singular = singular_values / torch.sum(singular_values)
        be_sum = normalize_singular * torch.log(normalize_singular)
        to_sum = torch.nan_to_num(be_sum, 0)
        neighbor_k =  -torch.sum(to_sum) / torch.log(torch.tensor(c))
        neighbor_homo += neighbor_k*nei_distri_label.size(0)
    neighbor_homo = neighbor_homo / data.num_nodes
    return neighbor_homo.item()
    
    

for data_name in datasets:
    dataset, data = load_dataset(data_name)
    c = dataset.num_classes
    homo = node_homophily(data)
    e_hom = edge_homophily(data)
    nei_homo = neighbor_homophily(data, c)
    print(data_name, homo, e_hom, nei_homo)
    
```
